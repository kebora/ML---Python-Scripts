```python
import pandas as pd
```

```python
datapath='C:/Users/user/Desktop/ML Practice/lab3.csv'
```

```python
df = pd.read_csv(datapath)
```

```python
df.head(5)
```

```python
# drop the columns that have categorical values and might not be important, except 'Country Name'
df = df.drop(labels=['Country Code','Indicator Name','Indicator Code'],axis=1)
df
```

```python
# convert Country Name to numerical format
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False)
encoded_country_names = encoder.fit_transform(df[['Country Name']])
encoded_country_names
```

```python
# convert encoded_country_names to a dataframe
ecn_df = pd.DataFrame(encoded_country_names)
ecn_df
```

```python
# drop the last column [2023-NaN]
# add the ecn_df to the original dataframe to form a new dataframe
df = pd.concat([ecn_df, df.iloc[:, 1:-1]], axis=1)
df
```

```python
# separate features and target
features = df.iloc[:, 1:-1]
target = df['2022']
target.info()
```

```python
# using train_test_split to split the data
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=0.33,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# import libraries for the model
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(units=8,activation='relu',input_shape=[327]),
    layers.Dense(units=4,activation='relu'),
    layers.Dense(units=4,activation='relu'),
    layers.Dense(units=1)
])
```

```python
#This is the value I used for the models' input_shape
features.columns.size
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(min_delta=0.001,patience=20,restore_best_weights=True)
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=1000,callbacks=[early_stopping],verbose=1)
```

```python
history_df = pd.DataFrame(history.history)
history_df['loss'].plot()
```
